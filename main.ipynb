```python
import tkinter as tk
from playsound import playsound
import random
import math
import numpy as np
import sys
from greedyAstar import *
# from astar import *  #uncomment for distance based algorithm
from Bot3 import *
import pandas as pd
from PIL import Image, ImageTk
import multiprocessing as mp
from functools import partial
import time
from Charger import *

class Counter:
    def __init__(self,canvas):
        self.avgHealth = 0
        self.canvas = canvas
        self.collisions=0
        self.damagedItems=0
        self.incompleteTasks=0
        self.chargersUsed=0
        self.completedTasks=0
#         self.canvas.create_text(70,50,text="Dirt collected: "+str(self.dirtCollected),tags="counter")
#         self.canvas.create_text(70,50,text="Damaged items: "+str(self.collisions),tags="counter")
#     def itemCollected(self, canvas):
#         self.dirtCollected +=1
#         self.canvas.itemconfigure("counter",text="Dirt collected: "+str(self.dirtCollected))
    
#     def collisions(self, canvas):
#         self.collisions +=1
#         self.canvas.itemconfigure("counter",text="Damaged Items: "+str(self.collisions))


class Task:
    def __init__(self,Id,source,destination,progress):
        self.id = Id
        self.source=source
        self.destination=destination
        self.progress=progress
        self.position=source
        self.name="T"+str(Id)
    
    def draw(self,canvas):
        canvas.create_rectangle(self.position[0]-10,self.position[1]+5, \
                                  self.position[0]+10,self.position[1]+10,outline="black", fill="yellow",tags=self.name)
#         self.attributes('-alpha',0.5)

    def getLocation(self):
        return self.source[0],self.source[1]
    
class Hub:
    def __init__(self,namep,xp,yp,purpose):
        self.centreX = xp
        self.centreY = yp
        self.name = namep
        self.purpose=purpose
        if namep in ("R1","R2","R3","R4","R5","R6","R7","R8"):
            self.doorX = xp+30
            self.doorY = yp
        else:
            self.doorX = xp-30
            self.doorY = yp
    def draw(self,canvas):
        if self.purpose=="K":
            canvas.create_rectangle(self.centreX-25,self.centreY-25, \
                                  self.centreX+25,self.centreY+25,outline="dark blue", fill="orange")
        elif self.purpose=="B":
            canvas.create_rectangle(self.centreX-25,self.centreY-25, \
                                  self.centreX+25,self.centreY+25,outline="dark blue", fill="magenta")
        else:
            canvas.create_rectangle(self.centreX-25,self.centreY-25, \
                                  self.centreX+25,self.centreY+25,outline="dark blue", fill="blue")
            
        body = canvas.create_rectangle(self.centreX-15,self.centreY-15, \
                                  self.centreX+15,self.centreY+15, \
                                  fill="white",tags=self.name)
        canvas.create_text(self.centreX,self.centreY,text=str(self.name),tags=self.name)

    def getLocation(self):
        return self.centreX, self.centreY


class Dirt:
    def __init__(self,namep):
        self.centreX = random.randint(250,650)
        self.centreY = random.randint(60,800)
        self.name = namep

    def draw(self,canvas):
        body = canvas.create_oval(self.centreX-1,self.centreY-1, \
                                  self.centreX+1,self.centreY+1, \
                                  fill="grey",tags=self.name)

    def getLocation(self):
        return self.centreX, self.centreY
    
def buttonClicked(x,y,registryActives):
    for rr in registryActives:
        if isinstance(rr,Bot):
            rr.x = x
            rr.y = y

def initialise(window):
    window.resizable(False,False)
    canvas = tk.Canvas(window,width=1000,height=1000)
    canvas.pack()
    canvas.create_rectangle(200,0,800,50,outline="dark blue", fill="yellow")
    canvas.create_text(500,25,text="Reception",fill="black")
    return canvas

def register(canvas,noOfBots):
    registryActives = []
    registryPassives = []
    registryTasks=[]
    noOfDirt = 300
    
    #emergence at reception
    for i in range(0,noOfBots):
        centreX=200 +(i+1)*100
        centreY=100  #+(i+1)*70
        bot = Bot("B"+str(i+1),canvas,centreX,centreY)
        registryActives.append(bot)
        bot.draw(canvas)



    #emergence at corners
#     centreX=100
#     centreY=100
#     bot = Bot("B1",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)
    
#     centreX=800
#     centreY=100
#     bot = Bot("B2",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)
    
#     centreX=800
#     centreY=600
#     bot = Bot("B3",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)
    
#     centreX=100
#     centreY=600
#     bot = Bot("B4",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)

#     centreX=300
#     centreY=600
#     bot = Bot("B5",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)
    
    
    #central area emergence
#     centreX=500
#     centreY=200
#     bot = Bot("B1",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)
    
#     centreX=500
#     centreY=300
#     bot = Bot("B2",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)
    
#     centreX=500
#     centreY=400
#     bot = Bot("B3",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)
    
#     centreX=500
#     centreY=500
#     bot = Bot("B4",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)

#     centreX=500
#     centreY=600
#     bot = Bot("B5",canvas,centreX,centreY)
#     registryActives.append(bot)
#     bot.draw(canvas)

    
    
    charger1 = Charger("Charger1",200,50,0)
    registryPassives.append(charger1)
    charger1.draw(canvas)
    
    charger2 = Charger("Charger2",500,50,0)
    registryPassives.append(charger2)
    charger2.draw(canvas)
    
    charger3 = Charger("Charger3",800,50,0)
    registryPassives.append(charger3)
    charger3.draw(canvas)
    
    list_of_rooms=[]
    for i in range (1,9):
        hub1 = Hub("R"+str(i),100,100+(i-1)*70,"R")
        registryPassives.append(hub1)
        hub1.draw(canvas)
        list_of_rooms.append(hub1.name)
    
    for i in range (1,7):
        hub1 = Hub("R"+str(i+8),900,100+(i-1)*70,"R")
        registryPassives.append(hub1)
        hub1.draw(canvas)
        list_of_rooms.append(hub1.name)
    
    Bedding1 = Hub("B"+str(1),900,520,"B")
    registryPassives.append(Bedding1)
    Bedding1.draw(canvas)
    list_of_rooms.append(Bedding1.name)
    
    Kitchen1 = Hub("K"+str(1),900,590,"K")
    registryPassives.append(Kitchen1)
    Kitchen1.draw(canvas)
    list_of_rooms.append(Kitchen1.name)
    
    temp_demand=[]
    df = pd.read_excel (r'Tasks.xlsx')
    df.columns=["Id","Source","Destination","Progress"];
    # IMP- later add code for inserting missing values in database for noise 
    total_tasks=0
    for index,row in df.iterrows():
        temp_demand.append([row["Id"],str(row["Source"]),str(row["Destination"]),row["Progress"]])
        total_tasks+=1
#     temp_demand=[[1,"R1","R15",0],[2,"R2","R12",0],[3,"R3","R10",0],[4,"R4","R11",0]]
#     temp_demand=[]
    for dem in temp_demand:
        if dem[1] in list_of_rooms and dem[2] in list_of_rooms:
            for t in registryPassives:
                if t.name==dem[1]:
                    dem[1]=(t.doorX,t.doorY)
                if t.name==dem[2]:
                    dem[2]=(t.doorX,t.doorY)
            task=Task(dem[0],dem[1],dem[2],dem[3])
            registryTasks.append(task)
            task.draw(canvas)

    collisions = Counter(canvas)
    canvas.bind( "<Button-1>", lambda event: buttonClicked(event.x,event.y,registryActives) )
    return registryActives, registryPassives, collisions ,registryTasks

    
def forPath(canvas,rr,registryActives,registryPassives,demand,destination,collisions):
    diffDir=True
    if destination:
        rr.x=destination[0]
        rr.y=destination[1]
        rr.vr=3.0
        rr.vl=3.0
        collisions=rr.getAwayFromObstacle(0.7,registryActives,collisions)
        
        if not rr.collision: checkDirection(rr,destination,registryPassives)
        else: rr.transferFunction(None,None)        
        rr.move(canvas,registryPassives,1.0,demand,destination,collisions)
#         rr.transferFunction()
    else:
        collisions=rr.getAwayFromObstacle(0.7,registryActives,collisions)
        chargerIntensityL, chargerIntensityR = rr.senseCharger(registryPassives)
        rr.transferFunction(chargerIntensityL, chargerIntensityR)
        rr.move(canvas,registryPassives,1.0,None,None,collisions)

def moveIt(canvas,registryActives,registryPassives,collisions,moves,window,registryTasks,path,source,randomMove):
    moves += 1
    totalCompletedTasks=0
    batteryRemaining=0
    for rr in registryActives:                           # check the movement of all the number of bots in the hospital
        if rr.battery>4000:
            rr.charging=False
        if rr.battery<=0 and not rr.collision:
            rr.temp_path=[]
        if not rr.task and registryTasks and rr.battery>=1000 and not rr.charging:                # check the tasks assigned to the bot and the tasks remaining in the registry
            demand2=[]
            for task in registryTasks:                   #for each remaining task
                taskSourceDist = rr.distanceTo(task)     #the distance of the bot from the source of each task
                demand2.append([task,taskSourceDist])
            demand2.sort(key = lambda x: x[1])
            rr.task.append(demand2[0][0])                #assign the task which has the source nearest to the bot
#             rr.task.append(registryTasks[0])
            rr.totalTasksCompleted+=1
            registryTasks.remove(rr.task[0])             #remove the task from the registry
        
        if ((rr.task) or (rr.charging or rr.battery>1000)) and rr.temp_path:                     #if task is assigned and path is also given 
#             if rr.atCharger and rr.source: 
#                 rr.temp_path=rr.findAPath(registryPassives,rr.task[0].destination,True)
#             rr.atCharger=False
            forPath(canvas,rr,registryActives,registryPassives,None,rr.temp_path[0],collisions) #update the position of the bot
            if rr.temp_path:
                rr.temp_path.pop(0)          #remove the point from the path points list
                if not rr.temp_path and rr.charging:
                    rr.charging=False
            if (not rr.temp_path) and rr.task and not rr.collision and not rr.charging:           #path is completed 
                if not rr.source and not rr.collision:                        #the bot has covered the source point
                    rr.source=True
                    canvas.delete(rr.task[0].name)
                    rr.itemPicked=True
                elif not rr.collision:                                    #the bot has reached the destination 
                    rr.task=[]                           #so remove the task assigned to the bot
                    rr.source=False                      #as bot has reached the destination
                    rr.temp_path=[]
                    rr.itemPicked=False
                    collisions.completedTasks+=1
                    randomMove+=1

        elif not rr.temp_path and not rr.collision:     #if task is assigned but path is unknown
            if rr.battery<1200:
                rr.temp_path=rr.findAPath(registryPassives,None,True)
                rr.charging=True
            elif rr.task and rr.source:
                rr.temp_path=rr.findAPath(registryPassives,rr.task[0].destination,True)
            elif rr.task: 
                rr.temp_path=rr.findAPath(registryPassives,rr.task[0].source,False) #find path to reach the destination of the task
        
        if (not rr.task and not registryTasks) or rr.collision:
            collisions=rr.getAwayFromObstacle(0.7,registryActives,collisions)
            rr.transferFunction(None,None)
            rr.move(canvas,registryPassives,1.0,None,None,collisions)

        numberOfMoves = 5000
        totalCompletedTasks+=rr.totalTasksCompleted
        batteryRemaining+=rr.battery
        
#         if moves>numberOfMoves or battery==0:
#             window.destroy()
#     print(randomMove)
#     print(totalCompletedTasks)
    if (totalCompletedTasks)==16 or moves>20000 or batteryRemaining==0:
        num_bots=0
        avgHealth=0
        totalCompletedTasks=0
        for rr in registryActives:
            avgHealth+=rr.health
            num_bots+=1
            collisions.damagedItems=collisions.damagedItems+len(rr.damagedItems)
            totalCompletedTasks+=rr.totalTasksCompleted    
        avgHealth=avgHealth/num_bots
        collisions.avgHealth=avgHealth
        collisions.incompleteTasks=20-totalCompletedTasks
        window.destroy()
    canvas.after(3,moveIt,canvas,registryActives,registryPassives,collisions,moves,window,registryTasks,rr.temp_path,rr.source,randomMove)
    return collisions

def runOneExperiment(noOfBots):
    window = tk.Tk()
    canvas = initialise(window)
    registryActives, registryPassives, collisions, registryTasks = register(canvas,noOfBots)
    moves = 0
    finalpath=[]
#     collisions=0
    randomMove=0
    gridMaker()
    start_time = time.time()
    moveIt(canvas,registryActives,registryPassives, collisions, moves,window,registryTasks,finalpath,False,randomMove)
    window.mainloop()
    end_time= time.time()
    print("{:.2f}".format(end_time-start_time)," ",collisions.collisions,"         ","{:.2f}".format(collisions.avgHealth),"\
    ",collisions.damagedItems,"            ",collisions.incompleteTasks,"            ",collisions.completedTasks,"       ",collisions.chargersUsed)
#     return collisions


print("Time    Collisions     Health      Damaged     Incomplete      completed    Chargers")

# for i in range(50):
runOneExperiment(3)
# sys.exc_clear()
```
